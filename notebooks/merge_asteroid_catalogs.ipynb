```python
import os
import pandas as pd
import numpy as np
from tess_asteroid_ml import *
```

```python
keep_cols = ['Object name',
             'Astrometric RA (hh:mm:ss)',
             "Astrometric Dec (dd mm'ss\")", 
             'Visual magnitude (V)', 
             'RA rate ("/h)', 
             'Dec rate ("/h)', 
             'ra', 
             'dec',
             'V_mag', 
             'name', 
             'id', 
             'Absolute magntiude (H)', 
             'Magnitude slope (G)',
             'Eccentricity', 
             'Perihelion (au)', 
             'Time of perihelion passage (JD)',
             'Longitude of ascending node (deg)', 
             'Argument of perihelion (deg)',
             'Inclination (deg)', 
             'Epoch (JD)', 
             'H_mag', 
             'kind',
            ]
```

```python
for sector in range(14, 27):
    for camera in range(1, 5):
        # print(f"{sector}-{camera}")
        
        cat = pd.read_csv(f"{os.path.dirname(PACKAGEDIR)}/data/jpl/jpl_small_bodies_tess_s{sector:04}-{camera}-0_catalog.csv", index_col=0)
        elem = pd.read_csv(f"{os.path.dirname(PACKAGEDIR)}/data/jpl/jpl_small_bodies_tess_s{sector:04}-{camera}-0_catalog_elem.csv", index_col=0)
        comet = pd.read_csv(f"{os.path.dirname(PACKAGEDIR)}/data/jpl/jpl_small_bodies_tess_s{sector:04}-{camera}-0_catalog_comet.csv", index_col=0)
        
        df = pd.merge(cat, elem, on=["Object name", "name", "id"], how="outer")
        df["kind"] = "a"
        mask = np.isin(df.id.values, comet.id.values)
        df["kind"][mask] = "c"
        
        # print(df.kind.value_counts())
        
        df.loc[:, keep_cols].to_csv(f"{os.path.dirname(PACKAGEDIR)}/data/jpl/merged/jpl_small_bodies_tess_s{sector:04}-{camera}-0_catalog.csv")
```
