```python
def projection_param(hemisphere):

    if hemisphere == 'south':
        cenlon = 90.
        cenlat = -90.#-66.560708333333#
        projection = ccrs.AzimuthalEquidistant(central_longitude=cenlon, central_latitude=cenlat)    
        wrap = False#True#
        title = "NASA TESS's View\nof the Southern\nHemisphere"
        min_lat, max_lat = -90, 0.
    
    elif hemisphere == 'north':
        cenlon = -90.
        cenlat = 90.#-66.560708333333#
        projection = ccrs.AzimuthalEquidistant(central_longitude=cenlon, central_latitude=cenlat)    
        wrap = False#
        title = "NASA TESS's View\nof the Northern\nHemisphere"
        min_lat, max_lat = 0., 90.
    
    elif hemisphere == 'both':
        cenlon = 0.
        cenlat = 0.
        galactic_coords = False
        if galactic_coords:
            cenlon += 180
        projection = ccrs.Mollweide(central_longitude=0)
        # projection = ccrs.PlateCarree(central_longitude=cenlon)
        wrap = True
        title = "NASA TESS's View\nof the Sky"
        min_lat, max_lat = -90, 90.

    return projection, cenlon, min_lat, max_lat, wrap
```

```python
img = imread('bluecmap.png')
pic = img[:, 0, :]
cmap = colors.LinearSegmentedColormap.from_list('NASA_blue', pic[::-1, :], N=pic.shape[0])
vmin = 150
vmax = 901.
cnorm = colors.LogNorm(vmin=vmin, vmax=vmax)
```

```python
nccds = 16
nsectors = 13
object_cum_count = []
sb_ephems_lowres = {}
track_color = {}

title = "NASA TESS's \nAll-Sky View"
credit = "By Jorge Martinez-Palomera \n     Veselin Kostov & Ethan Kruse"

for sec in range(0, 26):
    start = sec * nccds
    end   = start + nccds

    if sec == 0:
        sectxt = f"Sector 1 \n {secstarts[1]} - {secends[1]}"
    else:
        sectxt = f"Sector 1 - {sec + 1} \n {secstarts[1]} - {secends[sec + 1]}"
        
    fig = plt.figure(figsize=(70, 35))
    ax = plt.axes([0.0, 0.0, 1.0, 1.0], projection=projection)
    ax.spines['geo'].set_linewidth(0)
    for edgefile in edgefiles:
        elat, elon = np.loadtxt(edgefile, unpack=True)
        plt.scatter(elon, elat, c='w', alpha=0.01, zorder=-5,
                    marker='.', s=1, transform=ccrs.PlateCarree())
    # ax.set_xlim(-180, 180)
    # ax.set_ylim(-90, 90)
    ax.text(0.025, 0.96, title, transform=fig.transFigure,
             ha='left', va='top', multialignment='left',
             fontsize=100, fontweight="semibold")
    ax.text(0.025, 0.02, credit, transform=fig.transFigure,
             ha='left', va='bottom', multialignment='left',
             fontsize=70)
    ax.text(0.95, 0.02, sectxt, transform=fig.transFigure, 
            ha='right', va='bottom', multialignment='right',
            fontsize=90)
    for ccd in range(0, end):
        if ccd < 208:
            pkl = pkl1
        else:
            pkl = pkl2
            ccd -= 208
            
        lon = pkl["lon"][ccd]
        lat = pkl["lat"][ccd]
        sector_data = pkl["data"][ccd]
        
        lmin = (((cenlon - 178) + 180) % 360) - 180
        lmax = (((cenlon + 178) + 180) % 360) - 180
        wlon = ((lon - cenlon + 180) % 360) - 180
        
        if wrap and (lon.max() > lmax) and (lon.min() < lmin):
            bad = ((np.abs(wlon[:-1, :-1] - wlon[:-1, 1:]) > 355.) |
                (np.abs(wlon[:-1, :-1] - wlon[1:, :-1]) > 355.) |
                (np.abs(wlon[:-1, :-1] - wlon[1:, 1:]) > 355.))
            maskeddata = np.ma.masked_where(bad, sector_data)
            ax.pcolormesh(
                lon, lat, maskeddata, 
                cmap=cmap, norm=cnorm, alpha=1,
                transform=ccrs.PlateCarree(),
            )
        else:
            ax.pcolormesh(
                lon, lat, sector_data, 
                cmap=cmap, norm=cnorm, alpha=1,
                transform=ccrs.PlateCarree(),
            )

    if True:
        sec_ff = sorted(glob(f"{os.path.dirname(PACKAGEDIR)}/data/jpl/tracks/sector{sec + 1:04}/"
                              f"tess-ffi_s{sec + 1:04}-0-0_*_hires.feather"))
    
        for k, ff in tqdm(enumerate(sec_ff), total=len(sec_ff)):
            name = os.path.basename(ff)[19:-14]
            if name in sb_ephems_lowres.keys():
                continue
            else:
                aux = pd.read_feather(ff)
                # step = len(aux) // 27 - 1 if len(aux) > 100 else 4
                if aux.vmag.mean() < maglim:
                    icrs = SkyCoord(ra=aux.ra, dec=aux.dec, frame='icrs', unit='deg')
                    ecliptic = icrs.transform_to(BarycentricTrueEcliptic)
                    aux["lon"] = ecliptic.lon.value * 1
                    aux["lon"] -= 180.
                    aux["lon"] *= -1.
                    aux["lat"] = ecliptic.lat.value * 1
                    sb_ephems_lowres[name] = aux
                    if name not in track_color.keys():
                        track_color[name] = plt.cm.tab20(np.random.randint(0,20))
                else:
                    continue
        object_cum_count.append(len(sb_ephems_lowres))
    
        for name, track in sb_ephems_lowres.items():
            ax.scatter(track.lon, track.lat, marker=".", s=0.5, 
                       rasterized=True, color=track_color[name], transform=ccrs.PlateCarree())
            
        astertxt = f"{object_cum_count[-1]} Asteroids \nV < {maglim}"
        ax.text(0.95, 0.96, astertxt, transform=fig.transFigure, 
            ha='right', va='top', multialignment='right',
            fontsize=90)

    plt.savefig(f"./figures/animation/tess_all-sky_sector{sec+1:03}_tracks_mollweide.png", 
                transparent=False, format="png", bbox_inches="tight", dpi=300)
    plt.show()
    
    # break
```

```python
file_path = "https://tess.mit.edu/public/files/TESS_orbit_times.csv"
downlinks = pd.read_csv(file_path, comment="#")

# Sector 3 was used for testing ACS and downlink times are not accurate.
# Manual override using values from https://tess.mit.edu/public/files/TESS_FFI_observation_times.csv
downlinks.loc[
    np.logical_and(downlinks["Sector"] == 3, downlinks["Orbit"] == 13),
    ["Start of Orbit", "End of Orbit"],
] = ["2018-09-20 12:56:15", "2018-10-05 01:29:40"]
downlinks.loc[
    np.logical_and(downlinks["Sector"] == 3, downlinks["Orbit"] == 14),
    ["Start of Orbit", "End of Orbit"],
] = ["2018-10-05 03:29:40", "2018-10-17 21:17:58"]
```

```python
sec = 0
btjd0 = 2457000

sec_ff = sorted(glob(f"{os.path.dirname(PACKAGEDIR)}/data/jpl/tracks/sector{sec + 1:04}/"
                              f"tess-ffi_s{sec + 1:04}-0-0_*_hires.feather"))

df = downlinks.query(f"Sector == {sec + 1}")
tstart = Time(df["Start of Orbit"].values[0], scale="isot", format="jd")
tstop = Time(df["End of Orbit"].values[-1], scale="isot", format="jd")
time_array = np.arange(tstart.jd, tstop.jd, 2)

pred_pos_lat = pd.DataFrame(columns=time_array)
pred_pos_lon = pd.DataFrame(columns=time_array)
    
for k, ff in tqdm(enumerate(sec_ff), total=len(sec_ff)):
    name = os.path.basename(ff)[19:-14]
    if name in sb_ephems_lowres.keys():
        continue
    else:
        aux = pd.read_feather(ff)
        # step = len(aux) // 27 - 1 if len(aux) > 100 else 4
        if aux.vmag.mean() < maglim:
            icrs = SkyCoord(ra=aux.ra, dec=aux.dec, frame='icrs', unit='deg')
            ecliptic = icrs.transform_to(BarycentricTrueEcliptic)
            lat = ecliptic.lat.value * 1
            lon = ecliptic.lon.value * 1
            lon -= 180.
            lon *= -1.
            tr_lat = np.interp(time_array, aux.time + btjd0, lat, left=np.nan, right=np.nan)
            tr_lon = np.interp(time_array, aux.time + btjd0, lon, left=np.nan, right=np.nan)
            pred_pos_lat.loc[name] = tr_lat
            pred_pos_lon.loc[name] = tr_lon
            if name not in track_color1keys():
                track_color[name] = plt.cm.tab20(np.random.randint(0,20))
        else:
            continue
object_cum_count.append(len(sb_ephems_lowres))
color_array = [track_color[name] for name in track_color.keys() if name in pred_pos_lat.index.values]

fig = plt.figure(figsize=(70, 35))
ax = plt.axes([0.0, 0.0, 1.0, 1.0], projection=projection)
ax.spines['geo'].set_linewidth(0)

for tdx in range(len(time_array)):
    ax.plot(
        pred_pos_lon.values[:, tdx], 
        pred_pos_lat.values[:, tdx], 
        c=color_array,
        lw=1, 
        rasterized=True, 
        transform=ccrs.PlateCarree(),
    )
    plt.savefig(f"./figures/animation/tess_all-sky_sector{sec+1:03}_tracks_test.png", 
                transparent=False, format="png", bbox_inches="tight", dpi=300)
```
