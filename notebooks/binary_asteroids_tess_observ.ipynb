```python
observ2 = []
```

```python
for k, row in tqdm(bin_list.iterrows(), total=len(bin_list), disable=False):
    # if row['nid'] <= 119067:
    #     continue
    if np.isfinite(row['nid']):
        target = f"DES = 2{int(row['nid']):07}"
    else:
        target = row['prov_desig']
    # if target in set([x.target.unique()[0] for x in observ if len(x)>0]):
    #     continue
    df = target_observability(target=target)
    df["target"] = target
    observ2.append(df)

observ2 = pd.concat(observ2, axis=0)
```

```python
observ2 = observ2.reset_index(drop=True)
observ2.sector = observ2.sector.astype(int)
observ2.camera = observ2.camera.astype(int)
observ2.ccd = observ2.ccd.astype(int)
observ2.target = [int(x[7:]) if "DES" in x else x for x in observ2.target.values]
observ2
```
