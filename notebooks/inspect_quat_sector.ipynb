```python
import pandas as pd
import numpy as np
import lightkurve as lk
import matplotlib.pyplot as plt
from astropy.stats import sigma_clip
```

```python
sector = 2
camera = 1

ff = (f"{PACKAGEDIR}/data/engineering/TESSVectors_S1-26_FFI"
      f"/TessVectors_S{sector:03}_C{camera}_FFI.csv"
)
vectors = pd.read_csv(ff, skiprows=44)

mask1 = sigma_clip(vectors.Quat1_Med, masked=True)
mask2 = sigma_clip(vectors.Quat2_Med, masked=True)
mask3 = sigma_clip(vectors.Quat3_Med, masked=True)

print(mask1.mask.sum() / len(mask1), mask2.mask.sum() / len(mask1), mask3.mask.sum() / len(mask1))

plt.title(f"S {sector} Cam {camera}")
plt.plot(vectors.MidTime[~mask1.mask], vectors.Quat1_Med[~mask1.mask]*3600)
plt.plot(vectors.MidTime[~mask2.mask], vectors.Quat2_Med[~mask2.mask]*3600)
plt.plot(vectors.MidTime[~mask3.mask], vectors.Quat3_Med[~mask3.mask]*3600)
plt.show()

plt.title(f"S {sector} Cam {camera}")
plt.plot(vectors.MidTime[~mask1.mask], vectors.Quat1_StdDev[~mask1.mask]*3600)
plt.plot(vectors.MidTime[~mask2.mask], vectors.Quat2_StdDev[~mask2.mask]*3600)
plt.plot(vectors.MidTime[~mask3.mask], vectors.Quat3_StdDev[~mask3.mask]*3600)
plt.show()

cbvs = lk.correctors.load_tess_cbvs(
            sector=sector,
            camera=camera,
            ccd=4,
            cbv_type="MultiScale",
            band=2,
        )

plt.plot(cbvs.time.value, cbvs["VECTOR_1"])
plt.plot(cbvs.time.value, cbvs["VECTOR_2"])
plt.plot(cbvs.time.value, cbvs["VECTOR_3"])
plt.plot(cbvs.time.value, cbvs["VECTOR_4"])
plt.show()
```
